## Генераторы

```python
def squared(a):
    for x in a:
        yield x ** 2

a = list(range(10))
for elem in squared(a):
    print(elem, end=' ')
```

```python
func = squared
func
generator = func(a)
generator
next(generator, None)
```

```python
while True:
    print(next(generator), end=' ')
```

```python
generator = func(a)
while True:
    try:
        elem = next(generator)
        # loop body
        print(elem, end=' ')
    except StopIteration:
        break
```

```python
def yielder3():
    for i in range(10):
        yield i

def yielder2():
    for elem in yielder3():
        yield elem

def yielder1():
    yield from yielder2()  # Передает управление второму генератору, 
                           # пока он не исчерпается

for x in yielder1():
    print(x, end=' ')
```

```python
def cumsum():
    csum = 0.0
    while True:
        csum += yield csum  # Вообще говоря это уже корутина

arr = [1.0 + x / 2 for x in range(10)]
arr
gen = cumsum()
gen
gen.send(None) # equal to next(gen)
for elem in arr:
    print(gen.send(elem), end=' ')

for elem in gen:
    print("Can reach?")
```

```python
hw = 'hello world'
def hello():
    global hw  # можно явно указать и менять отсюда
    hw = 'broken, fix pls'
    print(hw)

hello()
hw

def hello2():
    global hw
    hw = 'fixed, dont worry'

hello2()
hw

def hello2():
    nonlocal hw
    hw = 'fixed, dont worry'
```

## Замыкания

```python
def outer(x):
    z = x
    def inner(y):
        nonlocal z
        return z * y
    return inner

multiply_by_3 = outer(3)
multiply_by_3(4)
outer(3)(4)
```

Функция тоже объект!

```python
multiply_by = [outer(x) for x in range(10)]
multiply_by[3](5)
multiply_by[6](10)
```

## Декораторы

```python
def salad(f):
    def internal(*args):
        print("salad")
        f(*args)
    return internal

def bread(f):
    def internal():
        print("bread")
        f()
        print("bread")
    return internal

def tomato(f):
    def internal():
        print("tomato")
        f()
    return internal

def f():
    print("beef")

sandwich = bread(salad(tomato(f)))
sandwich()
```

Так люди придумали декораторы

```python
@bread
@salad
@tomato
def g():
    print("chicken")

g()  
# то же самое что g = bread(salad(tomato(g)))
```

```python
g.__name__, g.__repr__()
```

Лучше не затирать эти поля

```python
from functools import wraps

def bread(h):
    #@wraps(h)
    def internal():
        print("bread")
        h()
        print("bread")
    return internal

@bread
def h():
    print("fish")
    
h()
h.__name__, h.__repr__()
```

```python
# function
def f():
    return 'f'
f.__call__()
f.__call__ = lambda: 'f new'
f.__call__()

# generator
def g():  # function
    yield 'g'
it = g.__call__()  # get generator
it.__next__()  # iterate

# coroutine
def h():
    s = yield 'h'
    yield s
it = h.__call__() # get coroutine
it.__next__() # first iteration
it.send('hey')
```
